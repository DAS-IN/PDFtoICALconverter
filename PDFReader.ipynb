```python
from PyPDF2 import PdfFileReader
import tabula
import pandas as pd
```

```python
pdf = "Stundenplan.PDF"

inputpdf = PdfFileReader(open(pdf, "rb"))
main_df = pd.DataFrame()

for i in range(inputpdf.numPages):
    df = pd.DataFrame(tabula.read_pdf(pdf,lattice=True,pages=i + 1,multiple_tables=True,pandas_options={"header": None},)[0],)
    df = df.replace({"\r": " "}, regex=True)
    main_df = main_df.append(df)

print(main_df)
```

```python
#Set index to consecutive numbers
main_df = main_df.reset_index()
```

```python
#remove first row
main_df = main_df.drop([0], axis = 0)
main_df
```

```python
#set first row to column name
main_df = main_df.rename(columns = main_df.iloc[0])
main_df = main_df.drop([1], axis = 0)
```

```python
#Set index to consecutive numbers
main_df = main_df.reset_index()
main_df
```

```python
#remove unecessary columns
main_df = main_df.drop(['index', 1], axis = 1)
```

```python
#duplicate the 'datum / Uhrzeit' column
main_df1 = main_df.copy(deep = True)
main_df1['Start_Time'] = main_df.loc[:,'Datum / Zeit']
main_df1
```

```python
#sort the columns in right order
main_df1 = main_df1[['Start_Time', 'Datum / Zeit','Bezeichnung', 'Raum', 'Dozent']]
```

```python
#rename 'Datum / Zeit' column
main_df1 = main_df1.rename(columns={"Datum / Zeit": "End_Time"})
```

```python
#transfer datatype object to string
main_df1['Start_Time'] = main_df1['Start_Time'].astype("string")
main_df1['End_Time'] = main_df1['End_Time'].astype("string")
main_df1
```

```python
#cut the endtime out of starttime
main_df2 = main_df1
main_df2['Start_Time'] = main_df1['Start_Time'].str[:16]
main_df2
```

```python
#cut the starttime out of endtime
main_df2.End_Time = main_df2.End_Time.str[0:10] + main_df2.End_Time.str[-6:]
main_df2
```

```python
#convert string to datetime64
main_df2['Start_Time'] =  pd.to_datetime(main_df2['Start_Time'], format='%d.%m.%Y %H:%M')
main_df2['End_Time'] =  pd.to_datetime(main_df2['End_Time'], format='%d.%m.%Y %H:%M')
```
